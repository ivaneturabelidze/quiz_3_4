```python
import pandas as pd
```

```python
# Read and display csv file
df = pd.read_csv('Earthquakes_database.csv')
df
```

```python
# Replace Latitude and Longitude columns with single column named Coordinates for ease of copying and searching the location
df['coordinates'] = df['Latitude'].astype(str) + ' ' + df['Longitude'].astype(str)
df = df.drop(columns=['Latitude', 'Longitude'])
df
```

```python
# Replace 'NaN' with 'No Info' so that it will be easier for people with less technical background to understand
df = df.fillna('No Info')
df
```

```python
# Categorize earthquakes by power of eathquake based on Magnitude
bins = [0, 4.0, 5.9, 6.9, 7.9, 10, float('inf')]
labels = ['Minor', 'Light', 'Moderate', 'Strong', 'Major', 'Extreme']
df['Earthquake Power'] = pd.cut(df['Magnitude'], bins=bins, labels=labels, include_lowest=True, right=False)
df
```

```python
import matplotlib.pyplot as plt
```

```python
df['Date'] = pd.to_datetime(df['Date'])
filtered_df = df[df['Date'] > '2010-01-01']
plt.plot(filtered_df['Date'], filtered_df['Magnitude'])
plt.show()
```

```python
source_count = df['Source'].value_counts()
sources_to_group = source_count[source_count < 100].index
df.loc[df['Source'].isin(sources_to_group), 'Source'] = 'Rest'
plt.figure(figsize=(8,8))
plt.pie(source_count, labels=source_count.index)
plt.show()
```
